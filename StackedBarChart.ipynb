```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

url = "https://drive.google.com/uc?export=download&id=1XQQsOKeTzzOulQPfY5WEhTK0rDCQ8rUf"
df = pd.read_csv(url)

def stacked_barchart(df, theme1, theme2, theme3, indicator):

    data = df[(df[theme1] == 1) | (df[theme2] == 1) | (df[theme3] == 1)]

    group = data.groupby([theme1, theme2, theme3])[indicator].value_counts()
    group.unstack().plot.bar(stacked=True)
    plt.legend(bbox_to_anchor=(1,0.5), loc="center right", fontsize=10, bbox_transform=plt.gcf().transFigure)
    plt.show()

stacked_barchart(df, 'Q1 support services talk', 'Q1 helpline', 'Q1 friends and family', 'gender-encoded')
```
