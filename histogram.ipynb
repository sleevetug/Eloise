```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as n

url = "https://drive.google.com/uc?export=download&id=144dlnCJ7rHs4LllVqXtsg2ABpnN_XucB"
df = pd.read_csv(url)

df = df[['Where are you based?','Which age range are you in?', 'How would you describe your gender?', 'Are you living with a disability or a long-term condition? If so how would you describe this?']]

def produce_histogram(column):

    data = df[column]

    cm = plt.cm.get_cmap('RdYlBu_r')

    plt.figure(figsize=(10,10))
    plt.ylabel('Number of People', fontsize=13)
    plt.xlabel('Age Range', fontsize=13)
    plt.xticks(rotation=45)
    n, bins, patches = plt.hist(data, 10, color='green', edgecolor='#343434')

    for i, p in enumerate(patches):
        plt.setp(p, 'facecolor', cm(i/10))

    plt.title('Participants Age Range', fontsize=15)
    plt.show()
    
produce_histogram('Which age range are you in?')
```
