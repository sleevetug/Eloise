```python
# Python function 2 applied to Gender column for the indicator _medway _in the Where are you based column with pie chart fully labelled and downloaded as .png

import pandas as pd
import matplotlib.pyplot as plt
url = "https://drive.google.com/uc?export=download&id=144dlnCJ7rHs4LllVqXtsg2ABpnN_XucB"
df = pd.read_csv(url)

df = df[['Where are you based?','Which age range are you in?', 'How would you describe your gender?', 'Are you living with a disability or a long-term condition? If so how would you describe this?']]

def encode_gender(df, **kwds):

      key = kwds['key']

      if df[key] == "Female":
        return df[key]
      if df[key] == "Male":
        return df[key]
      if df[key] == "Prefer not to say":
        return df[key]
      else:
        return "Other"

df["gender-encoded"] = df.apply(encode_gender, axis=1, key="How would you describe your gender?")

def produce_piechart(df, column1, column2, indicator):

    data = df[df[column2] == indicator][column1].value_counts()
    labels = data.keys()
    plt.figure(figsize=(10,10))
    colors = ['#f6b9ad', '#5e8d5a', '#f68f3c', '#ee6f68']
    plt.pie(data, labeldistance=1.15, wedgeprops = { 'linewidth' : 1, 'edgecolor' : 'white' }, colors=colors, autopct='%1.1f%%')
    plt.title("Gender of Participants Located in Medway", fontsize=15)
    plt.legend(labels=labels, bbox_to_anchor=(1,0.5), loc="center right", fontsize=10, bbox_transform=plt.gcf().transFigure)
    plt.show()

produce_piechart(df, 'gender-encoded', 'Where are you based?', 'Medway')
```

```python
# Python function 2 applied to Disability column for the indicator _medway _in the Where are you based column with pie chart fully labelled and downloaded as .png

import pandas as pd
import matplotlib.pyplot as plt

url = "https://drive.google.com/uc?export=download&id=144dlnCJ7rHs4LllVqXtsg2ABpnN_XucB"
df = pd.read_csv(url)

df = df[['Where are you based?', 'Which age range are you in?', 'How would you describe your gender?', 'Are you living with a disability or a long-term condition? If so how would you describe this?']]

def lowercase():
    df['Are you living with a disability or a long-term condition? If so how would you describe this?'] = df['Are you living with a disability or a long-term condition? If so how would you describe this?'].str.lower()
lowercase()

def encode_disability(df, **kwds):

    key = kwds['key']

    if df[key] == "no":
        return 1
    else:
        return 0

df["disability-encoded"] = df.apply(encode_disability, axis=1, key="Are you living with a disability or a long-term condition? If so how would you describe this?")

def produce_piechart(df, column1, column2, indicator):

    data = df[df[column2] == indicator][column1].value_counts()
    labels = ['Non-disabled', 'Disabled']
    plt.figure(figsize=(10,10))
    colors = ['#bbdfd5', '#e8cf37']
    plt.pie(data, labeldistance=1.15, wedgeprops = { 'linewidth' : 1, 'edgecolor' : 'white' }, colors=colors, autopct='%1.1f%%')
    plt.title("Disabled and Non-disabled Participants Located in Medway", fontsize=15)
    plt.legend(labels=labels, bbox_to_anchor=(1,0.5), loc="center right", fontsize=10, bbox_transform=plt.gcf().transFigure)
    plt.show()

produce_piechart(df, 'disability-encoded', 'Where are you based?', 'Medway')
```
